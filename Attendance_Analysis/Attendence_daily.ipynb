```python
intimeresult = [] 
for i,j in  zip(data[' In Time '],data[' Employee Code ']):
    if j==10013 or j==10030 or j==10031 or j==10035 or j==10036 or j==10041 or j==10042 or j==10043 or j==10050 or j==10051 or j==10052 or j==10041:
        
        if i.hour>=9 and i.minute >= 30:
            
            intimeresult.append("Late")
        elif i.hour>=7:
            intimeresult.append('Pass')
        elif i.hour==0:
             intimeresult.append("NoThumb")
        else:
            intimeresult.append(" ")
    else:
        if i.hour>=10 and i.minute >= 1:
            intimeresult.append("Late")
        elif i.hour>=7:
            intimeresult.append('Pass')
        elif i.hour==0:
             intimeresult.append("NoThumb")
        else:
            intimeresult.append(" ")
            
data["intimeResult"] = intimeresult 
```

```python
Day = [] 
for i in  data[" Duration "]:
    if i.hour>=9:
        Day.append("Present")
    elif i.hour == 4 and i.minute ==30 or i.hour ==8 or i.hour==6 or i.hour==5 or i.hour==7:
        Day.append("Halfday")
    else:
        Day.append("Absent")


        
        
data["Day"] = Day  
```

```python
data[' In Time '] = pd.to_datetime(data[' In Time ']).dt.time
data[' Duration '] = pd.to_datetime(data[' Duration ']).dt.time
```
