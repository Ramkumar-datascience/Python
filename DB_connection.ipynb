## SQLAlchemy

SQLAlchemy provides a nice “Pythonic” way of interacting with databases.

In order to be able to connect to the databases, we need to initiate something known as Connection Engine. This engine is dependent on the type of database that you are connecting to. It will be used to connect to the database engine when the script is executed. You can create the engine by using the following URI pattern

## Pandas DataFrame to SQL Table

```python
from pandas.io import sql
from sqlalchemy import create_engine
import pymysql
```

```python
engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                       .format(user="root",
                               pw="admin",
                               db="sakila"))
```

```python
data.to_sql(con=engine, name='mentors_house', if_exists='replace')
```

```python
df2 = pd.read_sql('SELECT * FROM mentors_house', con=engine)
df2
```

## SQL Table to DataFrame

```python
df1=pd.read_sql_table('mentors_house', engine)
df1
```
