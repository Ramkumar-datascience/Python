```python
driver = webdriver.Chrome(r'C:\Users\DELL\.wdm\drivers\chromedriver\win32\91.0.4472.101\chromedriver.exe')
driver.get("https://www.zomato.com/hyderabad/restaurants/biryani?category=1")
print(driver.title)
print(driver.current_url)
#print(driver.page_souurce)

for i in range(1000,10000,1000):
    driver.execute_script("window.scrollBy(0," +str(i)+")")
    time.sleep(1)


    html = driver.page_source
    soup = BeautifulSoup(html,"html.parser")
print(soup)
```

```python
data = soup.find_all('div',class_='sc-dwztqd AIgKo')
print(len(data))
names = []
ratings = []
price=[]
for i in data:
    name = i.find_all('p',class_='sc-1hez2tp-0 sc-hvvHee ifyMgS')
    #print(len(name))
    if len(name)<1:
        names.append('nan')
    else:
        names.append(name[0].text)
    rat = i.find_all('p',class_='sc-1hez2tp-0 lhdg1m-2 dgzDPB')
    #print(rat)
    if len(rat)<1:
        ratings.append('nan')
    else:
        ratings.append(rat[0].text)
    p = i.find_all('span',class_='sc-isBZXS fMwCee')
    if len(p)<1:
        price.append('nan')
    else:
        price.append(p[0].text)
```

```python
print(len(names))
print(len(ratings))
print(len(price))
```

```python
import pandas as pd
df = pd.DataFrame({'Name':names,
                  'Rating':ratings,
                  'Price':price})
```
